For each function equation, $d$ is desired number of parameters for the function (i.e. the dimensionality).

```python
from IPython.core.display import HTML
%pylab inline
```

```python
def plot_da_func(func, xlim, ylim, delta, labels=1):
    x = arange(xlim[0], xlim[1], delta)
    y = arange(ylim[0], ylim[1], delta)
    X, Y = meshgrid(x, y)
    Z = zeros([len(x), len(y)])
        
    for i, xx in enumerate(x):
        for j, yy in enumerate(y):
            if func is 'naive':
                Z[i,j] = naive(array([xx, yy]))
            elif func is 'ackley':
                Z[i,j] = ackley(array([xx, yy]))
            elif func is 'easom':
                Z[i,j] = easom(array([xx, yy]))
            elif func is 'rosenbrock':
                Z[i,j] = rosenbrock(array([xx, yy]))
            
    if labels is 1:
        CS = contour(X, Y, Z)
    else:
        CS = contour(X, Y, Z, labels)
        
    clabel(CS, inline=1, fontsize=10)
```

### A. My Function

$f(\mathbf{x}) = \sum\limits_{i=1}^d \cos(x_i) + 0.5\|\mathbf{x}\|_2 + d$

```python
def naive(x):
    return sum(-cos(x)) + 0.5*norm(x) + len(x)

figsize(10,10)
plot_da_func('naive', [-10, 10], [-10, 10], 0.1, [0.1, 1, 3, 4.5, 6, 9])
```

### B. The Ackley Function

$f(\mathbf{x}) = -a\exp \left (-b\sqrt{\frac{1}{d}\sum\limits_{i=1}^d x_i^2} \right) - \exp \left (\frac{1}{d} \sum\limits_{i=1}^d \cos(c x_i) \right ) + a + \exp(1)$

```python
def ackley(x):
    A = 20
    B = 0.2
    C = 2
    return -A * exp(-B*sqrt(sum(x**2)/len(x))) - exp(sum(cos(2*pi*x))/len(x)) + A + exp(1)

plot_da_func('ackley', [-10, 10], [-10, 10], 0.1)
```

### C. The Easom Function

$f(\textbf{x}) = -\left (\prod\limits_{i=1}^d \cos(x+\pi) \right ) \exp\left ( - \sum\limits_{i=1}^d x_i^2 \right ) + 1$


```python
def easom(x):
    prod = 1
    summ = 0
    for element in x:
        prod *= cos(element+pi)
    for element in x:
        summ += element*element
    return -prod*exp(-summ) + 1

plot_da_func('easom', [-10, 10], [-10, 10], 0.05, [0.1, 0.5, 0.9])
```

### D. The Rosenbrock Function

$f(\textbf{x}) = \sum\limits_{i=1}^{d-1} \left ( 100(x_{i+1} - x_i^2)^2 + (x_i-1)^2 \right )$

```python
def rosenbrock(x):
    summ = 0
    for i in range(len(x)-1):
        summ += 100*(x[i+1] + 1 - (x[i]+1)**2)**2 + (x[i])**2
    return summ
    
plot_da_func('rosenbrock', [-4, 2], [-4, 2], 0.01, [1, 10, 50, 100 ,200, 500, 1000, 5000])
```
